```python
import matplotlib.pyplot as plt
img = plt.imread("a.jpg")
plt.imshow(img)
```

```python
import cv2
import numpy as np
kernel = np.array([
    [1,1,1],
    [1,1,1],
    [1,1,1]
])
res = cv2.filter2D(img,-1,kernel)
plt.imshow(res)
```

```python
import cv2
import numpy as np
kernel = np.array([
    [0,0,0],
    [1,1,1],
    [0,0,0]
])
res = cv2.filter2D(img,-1,kernel)
plt.imshow(res)
```

```python
import cv2
import numpy as np
kernel = np.array([
    [0,1,0],
    [0,1,0],
    [0,1,0]
])
res = cv2.filter2D(img,-1,kernel)
plt.imshow(res)
```

```python
import cv2
import numpy as np
kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,0]
])
res = cv2.filter2D(img,-1,kernel)
plt.imshow(res)
```

```python
import cv2
import numpy as np
kernel = np.array([
    [1,1,0],
    [1,-4,1],
    [0,1,1]
])
res = cv2.filter2D(img,-1,kernel)
plt.imshow(res)
```
