```python
import numpy as np
X = np.loadtxt('test_image.csv',delimiter=',')
y = np.loadtxt('test_label.csv',delimiter=',').astype(int)
Y = np.eye(10)[y]
```

## 重要操作

```python
##把数据集拆分成训练集和测试集  80%的数据训练，20%的数据测试
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=23)
```

## 第一个模型的训练

```python
## 导包
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
import numpy as np

## 创建模型，指定输入和输出，指定激活函数
model = Sequential()
model.add(Dense(10,input_shape=(784,)))
model.add(Activation('softmax'))
model.summary()

## 编译模型，指定模型的训练参数
model.compile(optimizer='adam',loss="categorical_crossentropy",metrics=['accuracy'])

## 训练模型
data = model.fit(X_train,y_train,epochs=300,batch_size=128,validation_data=(X_test,y_test)) #在训练的过程中 就查看验证集的数据的准确度
```

```python
# data.history
```

```python
## 可视化训练集和验证集的损失函数的值， loss，val_loss
import matplotlib.pyplot as plt
loss_values = data.history['loss']
val_loss_values = data.history['val_loss']

line1 = plt.plot(np.arange(0,len(loss_values)),loss_values,label= "traing loss")
line2 = plt.plot(np.arange(0,len(loss_values)),val_loss_values,label = "validation  loss")

plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```

```python
## 可视化图表，把accuracy精确度的变换展示出来
import matplotlib.pyplot as plt
accuracy_values = data.history['accuracy']
val_accuracy_values = data.history['val_accuracy']

line1 = plt.plot(np.arange(0,len(accuracy_values)),accuracy_values,label= "traing accuracy")
line2 = plt.plot(np.arange(0,len(accuracy_values)),val_accuracy_values,label = "validation  accuracy")

plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```
